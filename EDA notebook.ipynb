```python
cab_data['price per KM'] = cab_data['Price Charged'] / cab_data['KM Travelled'] 
cab_data.head()
```

```python
price_df = cab_data.groupby(by= ['City']).sum()/ cab_data.groupby(by = ['City']).count()
price_df.sort_values('price per KM')
price_df.reset_index('City')
```

```python
print(price_df.columns.tolist())
new_price_df = pd.DataFrame({'City': price_df.index , 'price per KM' :price_df['price per KM'] })
```

```python
new_price_df.plot.bar()
```
